```python
def parseInput():
    filepath = 'day7_input'
    allParts = []
    with open(filepath) as file:
        # data = [line.rstrip() for line in file]
        data = file.read().rstrip()
        allParts = data.split(',')
        allParts = [ int(x) for x in allParts ]
    return allParts

```

### Part 1

```python
def ampProgram(phaseSetting, inputSignal, allInts):
    phaseSet = False
    remip = 0
    skipInc = False
    ip = 0
    
    lastOut = 0
    
    while ip < len(allInts):

        startint = str(allInts[ip])
        parammode = startint.rjust(5,'0')
        opcode = int(parammode[-2:])
        parammode = parammode[:-2]
        # reverse
        parammode = parammode[::-1]


        # print("**",ip,startint,opcode,parammode)

        if opcode == 99:
            return lastOut
        
        targetIndex = ip+1 if parammode[0] == "1" else allInts[ip+1]
        if opcode == 3:
            if not phaseSet:
                allInts[targetIndex] = phaseSetting
                phaseSet = True
            else:
                allInts[targetIndex] = inputSignal
            remip = 2
        elif opcode == 4:
            lastOut = allInts[targetIndex]
            print("signal to next amp:", lastOut)
            remip = 2

        else:
            param1 = allInts[ip+1]
            if parammode[0] == "0":
                param1 = allInts[param1]

            param2 = allInts[ip+2]
            if parammode[1] == "0":
                param2 = allInts[param2] 

            targetIndex = allInts[ip+3]

            if   opcode == 1:
                allInts[targetIndex] = param1 + param2
            elif opcode == 2:
                allInts[targetIndex] = param1 * param2
            elif opcode == 5:
                if param1 != 0:
                    ip = param2
                    skipInc = True
                else:
                    remip = 1
            elif opcode == 6:
                if param1 == 0:
                    ip = param2
                    skipInc = True
                else:
                    remip = 1
            elif opcode == 7:
                allInts[targetIndex] = 1 if param1 < param2 else 0
            elif opcode == 8:
                allInts[targetIndex] = 1 if param1 == param2 else 0

            elif opcode == 0:
                print("zero")
                break

        if not skipInc:
            ip += 4 -remip
            remip = 0
        else:
            skipInc = False
            
```

```python
def ampStepProgram(phaseSetting, inputSignal, allInts):
    phaseSet = True if phaseSetting < 0 else False
    remip = 0
    skipInc = False
    ip = inputSignal[1]
    
    lastOut = 0
    
    while ip < len(allInts):

        startint = str(allInts[ip])
        parammode = startint.rjust(5,'0')
        opcode = int(parammode[-2:])
        parammode = parammode[:-2]
        # reverse
        parammode = parammode[::-1]


        # print("**",ip,startint,opcode,parammode)

        if opcode == 99:
            return[lastOut,-1,allInts]
        
        targetIndex = ip+1 if parammode[0] == "1" else allInts[ip+1]
        if opcode == 3:
            if not phaseSet:
                allInts[targetIndex] = phaseSetting
                phaseSet = True
            else:
                allInts[targetIndex] = inputSignal[0]
            remip = 2
        elif opcode == 4:
            lastOut = allInts[targetIndex]
            return[lastOut,ip+2,allInts]
            remip = 2

        else:
            param1 = allInts[ip+1]
            if parammode[0] == "0":
                param1 = allInts[param1]

            param2 = allInts[ip+2]
            if parammode[1] == "0":
                param2 = allInts[param2] 

            targetIndex = allInts[ip+3]

            if   opcode == 1:
                allInts[targetIndex] = param1 + param2
            elif opcode == 2:
                allInts[targetIndex] = param1 * param2
            elif opcode == 5:
                if param1 != 0:
                    ip = param2
                    skipInc = True
                else:
                    remip = 1
            elif opcode == 6:
                if param1 == 0:
                    ip = param2
                    skipInc = True
                else:
                    remip = 1
            elif opcode == 7:
                allInts[targetIndex] = 1 if param1 < param2 else 0
            elif opcode == 8:
                allInts[targetIndex] = 1 if param1 == param2 else 0

            elif opcode == 0:
                print("zero")
                break

        if not skipInc:
            ip += 4 -remip
            remip = 0
        else:
            skipInc = False
            
```
