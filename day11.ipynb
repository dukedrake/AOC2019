```python
def parseInput():
    filepath = 'day11_input'
    allParts = []
    with open(filepath) as file:
        # data = [line.rstrip() for line in file]
        data = file.read().rstrip()
        allParts = data.split(',')
        allParts = [ int(x) for x in allParts ]
    return allParts

```

### Part 1

```python
class Scrubber:

    def __init__(self, allInts):
        self.allInts = { i : allInts[i] for i in range(0, len(allInts) ) }
        
        self.robotPosition = [0,0]
        self.robotDirection = 0
        self.colorPanels = {}
        
        self.updateStateState = False
        
        self.colorPanels[tuple(self.robotPosition)] = 0
        self.totalFields = 1
        self.totalSteps  = 1
        
        self.boostProgram()
        print("\n\nDONE",self.totalFields, self.totalSteps)
        
    
    def updateState(self, value):
        
        # update direction and move
        if self.updateStateState:
            self.robotDirection = (self.robotDirection + 1) % 4 if value else (self.robotDirection -1) % 4
            # print(self.robotDirection, end=",")
            
            # move
            if self.robotDirection % 2:
                if self.robotDirection == 1:
                    self.robotPosition[0] += 1
                else:
                    self.robotPosition[0] -= 1
            else:
                if self.robotDirection == 2:
                    self.robotPosition[1] += 1
                else:
                    self.robotPosition[1] -= 1
            
            
            # prepare target field
            rpTuple = tuple(self.robotPosition)
            if rpTuple not in self.colorPanels:
                self.colorPanels[rpTuple] = 0
                self.totalFields += 1
            self.totalSteps += 1
            
            # print(self.robotPosition, end=",")
            
            
        # paint panel
        else:
            self.colorPanels[tuple(self.robotPosition)] = value
                
        self.updateStateState = not self.updateStateState
        

    def getState(self):
        return self.colorPanels[tuple(self.robotPosition)]
        
        
    def boostProgram(self):
        skipInc = False
        ip = 0
        ipAdd = 0
        rb = 0
        lastOut = 0
        allInts = self.allInts

        while ip < len(allInts):

            startint = str(allInts[ip])
            parammode = startint.rjust(5,'0')
            opcode = int(parammode[-2:])
            parammode = parammode[:-2]
            # reverse
            parammode = parammode[::-1]


            # print("**",ip,opcode,'parammodes:',parammode,'relative base:',rb)

            ##################### exit opcode
            if opcode == 99:
                return lastOut

            ##################### one parameter opcodes
            ipAdd += 2
            targetIndex = allInts[ip+1]

            if parammode[0] == "2":
                targetIndex += rb


            ## input
            if opcode == 3:
                allInts[targetIndex] = self.getState()
                    
            ## output
            elif opcode == 4:
                outval = allInts[targetIndex]
                self.updateState(outval)

            ## change relative base
            elif opcode == 9:
                if parammode[0] == "1":
                    rb += allInts[ip+1]
                else:
                    if targetIndex not in allInts:
                        allInts[targetIndex] = 0
                    rb += allInts[targetIndex]


            ##################### two and three parameter opcodes
            else:
                ipAdd += 2

                param1 = allInts[ip+1]
                if parammode[0] == "0":
                    if param1 not in allInts:
                        allInts[param1] = 0
                    param1 = allInts[param1]
                elif parammode[0] == "2":
                    param1rb = param1+rb
                    if param1rb not in allInts:
                        allInts[param1rb] = 0
                    param1 = allInts[param1rb]

                param2 = allInts[ip+2]
                if parammode[1] == "0":
                    if param2 not in allInts:
                        allInts[param2] = 0
                    param2 = allInts[param2]
                elif parammode[1] == "2":
                    param2rb = param2+rb
                    if param2rb not in allInts:
                        allInts[param2rb] = 0
                    param2 = allInts[param2rb]

                if ip+3 not in allInts:
                    allInts[ip+3] = 0
                targetIndex = allInts[ip+3]
                if parammode[2] != "0":
                    targetIndex += rb


                # print("***",ip,opcode,'parammodes:',parammode,'relative base:',rb,param1,param2,targetIndex)

                if   opcode == 1:
                    allInts[targetIndex] = param1 + param2
                elif opcode == 2:
                    allInts[targetIndex] = param1 * param2
                elif opcode == 5:
                    if param1 != 0:
                        ip = param2
                        skipInc = True
                    else:
                        ipAdd -= 1
                elif opcode == 6:
                    if param1 == 0:
                        ip = param2
                        skipInc = True
                    else:
                        ipAdd -= 1
                elif opcode == 7:
                    allInts[targetIndex] = 1 if param1 < param2 else 0
                elif opcode == 8:
                    allInts[targetIndex] = 1 if param1 == param2 else 0

                elif opcode == 0:
                    print("zero")
                    break

            if not skipInc:
                ip += ipAdd
            else:
                skipInc = False
            ipAdd = 0

```

```python
allInts = parseInput()

myScrubber = Scrubber(allInts)

```
