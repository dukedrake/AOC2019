```python
def parseInput():
    filepath = 'day9_input'
    allParts = []
    with open(filepath) as file:
        # data = [line.rstrip() for line in file]
        data = file.read().rstrip()
        allParts = data.split(',')
        allParts = [ int(x) for x in allParts ]
    return allParts

```

### Part 1+2

```python
def boostProgram(inputSignal, allInts):
    skipInc = False
    ip = 0
    ipAdd = 0
    rb = 0
    lastOut = 0
    allInts = { i : allInts[i] for i in range(0, len(allInts) ) }
    
    while ip < len(allInts):

        startint = str(allInts[ip])
        parammode = startint.rjust(5,'0')
        opcode = int(parammode[-2:])
        parammode = parammode[:-2]
        # reverse
        parammode = parammode[::-1]


        # print("**",ip,opcode,'parammodes:',parammode,'relative base:',rb)

        ##################### exit opcode
        if opcode == 99:
            return lastOut
        
        ##################### one parameter opcodes
        ipAdd += 2
        targetIndex = allInts[ip+1]
        
        if parammode[0] == "2":
            targetIndex += rb
            
        
        ## input
        if opcode == 3:
            allInts[targetIndex] = inputSignal
        ## output
        elif opcode == 4:
            lastOut = allInts[targetIndex]
            
            print("  OUT  ",lastOut)
        ## change relative base
        elif opcode == 9:
            if parammode[0] == "1":
                rb += allInts[ip+1]
            else:
                if targetIndex not in allInts:
                    allInts[targetIndex] = 0
                rb += allInts[targetIndex]
            

        ##################### two and three parameter opcodes
        else:
            ipAdd += 2
            
            param1 = allInts[ip+1]
            if parammode[0] == "0":
                if param1 not in allInts:
                    allInts[param1] = 0
                param1 = allInts[param1]
            elif parammode[0] == "2":
                param1rb = param1+rb
                if param1rb not in allInts:
                    allInts[param1rb] = 0
                param1 = allInts[param1rb]

            param2 = allInts[ip+2]
            if parammode[1] == "0":
                if param2 not in allInts:
                    allInts[param2] = 0
                param2 = allInts[param2]
            elif parammode[1] == "2":
                param2rb = param2+rb
                if param2rb not in allInts:
                    allInts[param2rb] = 0
                param2 = allInts[param2rb]

            if ip+3 not in allInts:
                allInts[ip+3] = 0
            targetIndex = allInts[ip+3]
            if parammode[2] != "0":
                targetIndex += rb
            
            
            # print("***",ip,opcode,'parammodes:',parammode,'relative base:',rb,param1,param2,targetIndex)
            
            if   opcode == 1:
                allInts[targetIndex] = param1 + param2
            elif opcode == 2:
                allInts[targetIndex] = param1 * param2
            elif opcode == 5:
                if param1 != 0:
                    ip = param2
                    skipInc = True
                else:
                    ipAdd -= 1
            elif opcode == 6:
                if param1 == 0:
                    ip = param2
                    skipInc = True
                else:
                    ipAdd -= 1
            elif opcode == 7:
                allInts[targetIndex] = 1 if param1 < param2 else 0
            elif opcode == 8:
                allInts[targetIndex] = 1 if param1 == param2 else 0

            elif opcode == 0:
                print("zero")
                break

        if not skipInc:
            ip += ipAdd
        else:
            skipInc = False
        ipAdd = 0
            
```

```python
allInts = parseInput()
# allInts = [109,1,204,-1,1001,100,1,100,1008,100,16,101,1006,101,0,99]

boostProgram(2, allInts)
```
