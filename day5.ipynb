```python
def parseInput():
    filepath = 'day5_input'
    allParts = []
    with open(filepath) as file:
        # data = [line.rstrip() for line in file]
        data = file.read().rstrip()
        allParts = data.split(',')
        allParts = [ int(x) for x in allParts ]
    return allParts

'''
def isset(clist,cindex):
    try:
        clist[cindex]
    except IndexError:
        return False
    return True
'''
```

```python
allInts = parseInput()

rem2 = False
ip = -4
while ip < len(allInts):
    ip += 4
    if rem2:
        ip -= 2
        rem2 = False
        
    startint = str(allInts[ip])
    parammode = startint.rjust(5,'0')
    opcode = int(parammode[-2:])
    parammode = parammode[:-2]
    # reverse
    parammode = parammode[::-1]

    
    targetIndex = allInts[ip+1]
    # print("**",startint,opcode,parammode,targetIndex)

    
    if opcode == 99:
        break
    elif opcode == 3:
        allInts[allInts[ip+1]] = int(input())
        rem2 = True
        continue
    elif opcode == 4:
        print("out:",allInts[allInts[ip+1]])
        rem2 = True
        continue


    param1 = allInts[ip+1]
    if parammode[0] == "0":
        param1 = allInts[param1] 

    param2 = allInts[ip+2]
    if parammode[1] == "0":
        param2 = allInts[param2] 

    targetIndex = allInts[ip+3]

    if   opcode == 1:
        allInts[targetIndex] = param1 + param2
    elif opcode == 2:
        allInts[targetIndex] = param1 * param2

    elif opcode == 0:
        print("zero")
        break
            
```

```python
allInts = parseInput()

remip = 0
skipInc = False
ip = -4
while ip < len(allInts):
    if not skipInc:
        ip += 4
        ip -= remip
        remip = 0
    else:
        skipInc = False
    
        
    startint = str(allInts[ip])
    parammode = startint.rjust(5,'0')
    opcode = int(parammode[-2:])
    parammode = parammode[:-2]
    # reverse
    parammode = parammode[::-1]

    
    targetIndex = allInts[ip+1] if parammode[0] == "0" else ip+1
    # print("**",startint,opcode,parammode,targetIndex)
    
    
    if opcode == 99:
        break
    elif opcode == 3:
        allInts[targetIndex] = int(input())
        remip = 2
        continue
    elif opcode == 4:
        print("out:",allInts[targetIndex])
        remip = 2
        continue
    


    param1 = allInts[ip+1]
    if parammode[0] == "0":
        param1 = allInts[param1] 

    param2 = allInts[ip+2]
    if parammode[1] == "0":
        param2 = allInts[param2] 

    targetIndex = allInts[ip+3]

    if   opcode == 1:
        allInts[targetIndex] = param1 + param2
    elif opcode == 2:
        allInts[targetIndex] = param1 * param2
    elif opcode == 5:
        if param1 != 0:
            ip = param2
            skipInc = True
        else:
            remip = 1
    elif opcode == 6:
        if param1 == 0:
            ip = param2
            skipInc = True
        else:
            remip = 1
    elif opcode == 7:
        allInts[targetIndex] = 1 if param1 < param2 else 0
    elif opcode == 8:
        allInts[targetIndex] = 1 if param1 == param2 else 0

    elif opcode == 0:
        print("zero")
        break
            
```
