```python
def parseInput():
    filepath = 'day10_input'
    allParts = []
    with open(filepath) as file:
        data = [line.rstrip() for line in file]
    return data

```

### Part 1

```python
def findAstInQuadrant(xOff, yOff, xDelta, yDelta ):
    global astMap
    lineSlopes = {}
    lenMap  = len(astMap)
    lenLine = len(astMap[0])
    runLeft = runRight = xOff
    runUp = runDown = yOff
    
    tickcount = 500
    while runDown < lenMap and runDown >= 0:
        if tickcount < 0:
            break
        runRight = xOff
        # print(" "+str(runDown), end="  ")
        while runRight < lenLine and runRight >= 0:
            
            # exclude self
            if runRight == xOff and runDown == yOff:
                runRight += xDelta
                continue
        
            curItem = astMap[runDown][runRight]
            if curItem != '.' and curItem != "P":
                astMap[runDown][runRight] = "P"
                locRight = runRight - xOff
                locDown  = runDown  - yOff
                
                
                # locElemOut = str(runRight)+","+str(runDown)+'('+str(locRight)+","+str(locDown)+')'
                locElem = (runRight,runDown)
                if locRight != 0:
                    curSlope = locDown/locRight
                    # account for negative 0.0
                    if curSlope == 0.0 and locRight < 0:
                        curSlope = -0.00001
                # account for negative denominator
                else:
                    curSlope = -100.0
                
                # print(locElemOut,curSlope,end=',,')

                if curSlope not in lineSlopes:
                    lineSlopes[curSlope] = set()

                lineSlopes[curSlope].add(locElem)
            
                    
            # print(curItem+" ", end='')
            runRight += xDelta

        runDown += yDelta
            
    return lineSlopes
```

```python
starMap = parseInput()

# starMap = ["#.........", "...a......", "...b..a...", ".edcg....a", "..f.c.b...", ".....c....", "..efd.c.gb", ".......c..", "....f...c.", "...e..d..c"]
# starMap = ['.#..#', '.....', '#####', '....#', '...##']

# first example.  ((5,8), 33)
# starMap = ['......#.#.', '#..#.#....', '..#######.', '.#.#.###..', '.#..#.....', '..#....#.#', '#..#....#.', '.##.#..###', '##...#..#.', '.#....####']

# second example (1,2), 35
# starMap = ['#.#...#.#.', '.###....#.', '.#....#...', '##.#.#.#.#', '....#.#.#.', '.##..###.#', '..#...##..', '..##....##', '......#...', '.####.###.']

# big example (11,13), 210
# starMap = ['.#..##.###...#######', '##.############..##.', '.#.######.########.#', '.###.#######.####.#.', '#####.##.#.##.###.##', '..#####..#.#########', '####################', '#.####....###.#.#.##', '##.#################', '#####.##.###..####..', '..######..##.#######', '####.##.####...##..#', '.#####..#.######.###', '##...#.##########...', '#.##########.#######', '.####.#.###.###.#.##', '....##.##.###..#####', '.#.#.###########.###', '#.#.#.#####.####.###', '###.##.####.##.#..##']

lenMap  = len(starMap)
lenLine = len(starMap[0])

def transpose(m):
    return list(map(list, zip(*m)))

def getMap():
    astMap = []
    for y in range(lenMap):
        astMap.append([])
        for x in range(lenLine):
            astMap[y].append('.')
            mapCoord = starMap[y][x]
            if mapCoord == '#':
                astMap[y][-1] = 'X'
    return astMap





astList = {}
for y in range(lenMap):
    for x in range(lenLine):
        mapCoord = starMap[y][x]
        if mapCoord == '#':
            astList[(x,y)] = 'O'

maxAst = 0
maxLoc = ()
maxSlopes = {}
resMap = getMap()

for asteroid in astList:
    xOff = asteroid[0]
    yOff = asteroid[1]
        
    curSlopes = []
    astMap = getMap()
    astMap[yOff][xOff] = 'L'
    
    curSlopes.append(findAstInQuadrant(xOff, yOff,  1, -1))
    curSlopes.append(findAstInQuadrant(xOff, yOff,  1, 1))
    curSlopes.append(findAstInQuadrant(xOff, yOff, -1, 1))
    curSlopes.append(findAstInQuadrant(xOff, yOff, -1, -1))


    tlength = 0
    for c in curSlopes:
        # for slope in c:
            # print(slope, c[slope])
        tlength += len(c)
        # print("")

    resMap[yOff][xOff] = tlength
    if tlength > maxAst:
        maxAst = tlength
        maxLoc = (xOff,yOff)
        maxSlopes = curSlopes


        
print(maxAst)
print(maxLoc)

```
