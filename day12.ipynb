```python
import itertools

# testvalues
# moonPos = {"A":[-1, 0, 2], "B":[2,  -10, -7], "C":[4,  -8,  8], "D":[3,  5,   -1]}

moonPos = {"A":[19, -10, 7], "B":[1, 2, -3], "C":[14, -4, 1], "D":[8, 7, -6]}

moonVel = {"A":[0, 0, 0], "B":[0, 0, 0], "C":[0, 0, 0], "D":[0, 0, 0]}


```

```python


for i in range(1000):
    perms = itertools.combinations(moonPos,2)
    for moonPerm in perms:
        compA = moonPos[moonPerm[0]]
        compB = moonPos[moonPerm[1]]
        velA  = moonVel[moonPerm[0]]
        velB  = moonVel[moonPerm[1]]

        for j in range(3):
            if compA[j] > compB[j]:
                velA[j] -= 1
                velB[j] += 1
            elif compA[j] < compB[j]:
                velA[j] += 1
                velB[j] -= 1

    moonETotTot = 0
    for moonId in moonPos:
        moonEPot = 0
        moonEKin = 0
        for j in range(3):
            moonPos[moonId][j] += moonVel[moonId][j]
            moonEPot += abs(moonPos[moonId][j])
            moonEKin += abs(moonVel[moonId][j])
        moonETot = moonEPot * moonEKin
        if i == 999:
            print("ETot, ",moonId, moonETot)
        moonETotTot += moonETot

print(moonPos)
print(moonVel)
print(moonETotTot)
    
#163564, 163400 too high

#204 too low
```

```python
checkSumPos = {}
checkSumVel = {}

for i in range(30):
    perms = itertools.combinations(moonPos,2)
    for moonPerm in perms:
        compA = moonPos[moonPerm[0]]
        compB = moonPos[moonPerm[1]]
        velA  = moonVel[moonPerm[0]]
        velB  = moonVel[moonPerm[1]]

        for j in range(3):
            if compA[j] > compB[j]:
                velA[j] -= 1
                velB[j] += 1
            elif compA[j] < compB[j]:
                velA[j] += 1
                velB[j] -= 1

    moonETotTot = 0
    for moonId in moonPos:
        moonEPot = 0
        moonEKin = 0
        for j in range(3):
            moonPos[moonId][j] += moonVel[moonId][j]
            moonEPot += abs(moonPos[moonId][j])
            moonEKin += abs(moonVel[moonId][j])
        moonETot = moonEPot * moonEKin
        moonETotTot += moonETot
    
    if moonETotTot == 636:
        print("***",i,moonETotTot)
    
    print("Round",i)
    print("Positions", list(moonPos.items()))
    print("Velocities", list(moonVel.items()))
    
    if moonETotTot not in checkSumPos:
        checkSumPos[moonETotTot] = list(moonPos.items())
        checkSumVel[moonETotTot] = list(moonVel.items())
    else:
        print("Found checksum",i,moonETotTot)
        print("Positions with same Checksum:", checkSumPos[moonETotTot])
        print("Velocities with same Checksum:", checkSumVel[moonETotTot])
        
    print("")
        
        
        
```
