```python
def parseInput():
    filepath = 'day13_input'
    allParts = []
    with open(filepath) as file:
        # data = [line.rstrip() for line in file]
        data = file.read().rstrip()
        allParts = data.split(',')
        allParts = [ int(x) for x in allParts ]
    return allParts

```

### Part 1 + 2

```python
class Breakout:

    def __init__(self, allInts):
        self.allInts = { i : allInts[i] for i in range(0, len(allInts) ) }
        
        # part 2, play for free
        self.allInts[0] = 2
        self.totalRounds = 0
        
        self.screen = {}
                
        self.outTriple = []
        self.totalBlocks = 0
        self.ballX = 0
        self.paddleX = 0

        self.breakerProgram()
        print("\n\nDONE",self.totalRounds, "\n\n")
        print("Score:", self.screen[-1][0],"\n")
        
       
    
    def updateState(self, value):
        self.outTriple.append(value)
        if len(self.outTriple) == 3:
            if self.outTriple[0] not in self.screen:
                self.screen[self.outTriple[0]] = {}
            
            self.screen[self.outTriple[0]][self.outTriple[1]] = self.outTriple[2]
            if self.outTriple[2] == 3:
                self.paddleX = self.outTriple[0]
            if self.outTriple[2] == 4:
                self.ballX = self.outTriple[0]
            
            self.totalBlocks += 1
            self.outTriple = []
        
        

    def requestJoystickInput(self):
        self.totalRounds += 1
        # print("painted Blocks:",self.totalBlocks)
        # print("Score:", self.screen[-1][0],"\n")
        
        # self.drawScreen()
        
        if self.ballX > self.paddleX:
            return 1
        if self.ballX < self.paddleX:
            return -1
        return 0
    
    

    
    
    def drawScreen(self):
        leftMost = 0
        rightMost = -200
        topMost = 0
        bottomMost = -200
        for x in self.screen:
            if x > rightMost:
                rightMost = x
            
            for y in self.screen[x]:
                if y > bottomMost:
                    bottomMost = y
       
        
        for y in range(16,bottomMost+1):
            for x in range(leftMost,rightMost+1):
            
                out = '.'
                if x in self.screen and y in self.screen[x]:
                    if self.screen[x][y] == 1:
                        out = '#'
                    elif self.screen[x][y] == 2:
                        out = 'M'
                    elif self.screen[x][y] == 3:
                        out = '='
                    elif self.screen[x][y] == 4:
                        out = 'O'
                        
                print(out,end=" ")
            print("")
        
        
        
        
    def breakerProgram(self):
        skipInc = False
        ip = 0
        ipAdd = 0
        rb = 0
        lastOut = 0
        allInts = self.allInts

        while ip < len(allInts):

            startint = str(allInts[ip])
            parammode = startint.rjust(5,'0')
            opcode = int(parammode[-2:])
            parammode = parammode[:-2]
            # reverse
            parammode = parammode[::-1]


            # print("**",ip,opcode,'parammodes:',parammode,'relative base:',rb)

            ##################### exit opcode
            if opcode == 99:
                return lastOut

            ##################### one parameter opcodes
            ipAdd += 2
            targetIndex = allInts[ip+1]

            if parammode[0] == "2":
                targetIndex += rb


            ## input
            if opcode == 3:
                allInts[targetIndex] = self.requestJoystickInput()
                    
            ## output
            elif opcode == 4:
                if parammode[0] == "1":
                    outval = allInts[ip+1]
                else:
                    outval = allInts[targetIndex]
                self.updateState(outval)

            ## change relative base
            elif opcode == 9:
                if parammode[0] == "1":
                    rb += allInts[ip+1]
                else:
                    if targetIndex not in allInts:
                        allInts[targetIndex] = 0
                    rb += allInts[targetIndex]


            ##################### two and three parameter opcodes
            else:
                ipAdd += 2

                param1 = allInts[ip+1]
                if parammode[0] == "0":
                    if param1 not in allInts:
                        allInts[param1] = 0
                    param1 = allInts[param1]
                elif parammode[0] == "2":
                    param1rb = param1+rb
                    if param1rb not in allInts:
                        allInts[param1rb] = 0
                    param1 = allInts[param1rb]

                param2 = allInts[ip+2]
                if parammode[1] == "0":
                    if param2 not in allInts:
                        allInts[param2] = 0
                    param2 = allInts[param2]
                elif parammode[1] == "2":
                    param2rb = param2+rb
                    if param2rb not in allInts:
                        allInts[param2rb] = 0
                    param2 = allInts[param2rb]

                if ip+3 not in allInts:
                    allInts[ip+3] = 0
                targetIndex = allInts[ip+3]
                if parammode[2] != "0":
                    targetIndex += rb


                # print("***",ip,opcode,'parammodes:',parammode,'relative base:',rb,param1,param2,targetIndex)

                if   opcode == 1:
                    allInts[targetIndex] = param1 + param2
                elif opcode == 2:
                    allInts[targetIndex] = param1 * param2
                elif opcode == 5:
                    if param1 != 0:
                        ip = param2
                        skipInc = True
                    else:
                        ipAdd -= 1
                elif opcode == 6:
                    if param1 == 0:
                        ip = param2
                        skipInc = True
                    else:
                        ipAdd -= 1
                elif opcode == 7:
                    allInts[targetIndex] = 1 if param1 < param2 else 0
                elif opcode == 8:
                    allInts[targetIndex] = 1 if param1 == param2 else 0

                elif opcode == 0:
                    print("zero")
                    break

            if not skipInc:
                ip += ipAdd
            else:
                skipInc = False
            ipAdd = 0

```
